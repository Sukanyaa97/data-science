```
import pandas as pd

file_path = '/content/drive/My Drive/Data_science/simple_sales.csv'

df = pd.read_csv(file_path)  # อ่านไฟล์ CSV และสร้าง DataFrame

print(df)   # แสดง DataFrame
```
